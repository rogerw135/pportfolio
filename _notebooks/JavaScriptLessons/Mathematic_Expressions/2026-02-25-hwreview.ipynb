```python
%%js 
class Dog {
  constructor(name, breed) {
    this.name = name;
    this.breed = breed;
  }

  bark() {
    console.log(this.name + " says: Woof!");
  }

  describe() {
    console.log(this.name + " is a " + this.breed + ".");
  }
}

// Create an object from the class
let myDog = new Dog("Buddy", "Golden Retriever");

// Call the methods
myDog.bark();
myDog.describe();
```
